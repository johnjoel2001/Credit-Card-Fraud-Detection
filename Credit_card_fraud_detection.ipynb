```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
legit=df[df.Class==0]
legit['Amount'].describe()
```

```python
fraud=df[df.Class==1]
fraud['Amount'].describe()
```

```python
df.groupby('Class').mean()
```

```python
df['Class'].value_counts()
```

### We see that the datset is imbalanced.
<B>We will use undersampling technique.

```python
legit_under=legit.sample(492)
legit_under.shape
```

```python
fraud.shape
```

```python
new_df=pd.concat([legit_under,fraud],axis=0)
new_df.head()
```

```python
new_df.shape
```

```python
X=new_df.drop('Class',axis=1)
Y=new_df['Class']
```

### Now we have a balanced dataset

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
st=MinMaxScaler()
X['Amount']=st.fit_transform(X['Amount'].values.reshape(-1,1))
```

```python
X.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

model_params={
    'svm':{
        'model':SVC(gamma='auto'),
         'params':{
             'C':[1,10,20],
             'kernel':['rbf','linear']
         }
    },
    'randomforest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'logistic regression':{
        'model':LogisticRegression(),
        'params':{
            'C':[1,10,20]
        }
    }
}
```

```python
model_params
```

```python
lm=LogisticRegression()
lm.fit(X_train,Y_train)
```

```python
lm.score(X_train,Y_train)
```

```python
scores=[]
for mod_name,paramets in model_params.items():
    clf=GridSearchCV(paramets['model'],paramets['params'],cv=5,return_train_score=False)
    clf.fit(X_train,Y_train)
    
    scores.append({
        'model': mod_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

<b> We observe that random forest yields a beeter score than svm and log_regression
    

```python
rv=RandomForestClassifier()
rv.fit(X_train,Y_train)
```

```python
ypred=rv.predict(X_test)
```

```python
ypred[:5]
```

```python
Y_test[:5]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
plt.figure(figsize=(10,7))
print(classification_report(Y_test,ypred))
```

```python
cm=confusion_matrix(Y_test,ypred)
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted',weight='bold')
plt.ylabel('Actual',weight='bold')
```
